```python
import pandas as pd #reading csv
import numpy as np # linear regression
```

```python
#reading csv files
data1=pd.read_csv("/content/mental-and-substance-use-as-share-of-disease.csv")
data2=pd.read_csv("/content/prevalence-by-mental-and-substance-use-disorder.csv")
```

```python
#for ploting and representation
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Review the data
data1.head(5)
```

```python
data2.head(5)
```

```python
# combining data
data=pd.merge(data2,data1)
data.head(5)
```

```python
#data cleaning
data.isnull().sum()
```

```python
#deleting useless
data.drop('Code',inplace=True,axis=1)
```

```python
#view data
data.head(5)
```

```python
#renaming column
data.set_axis(['Country','Year','Schizophrenia','Bipolar','Eating disorders','Anxiety','Drug use','Depression','Alcohol use','Mental_fitness'], axis='columns', inplace=True)
```

```python
#heatmap
sns.heatmap(data.corr(),annot=True,cmap='Greens')
plt.plot()
```

```python
#pair graphs
sns.pairplot(data,corner=True)
plt.show()
```

```python
#mean
mean=data['Mental_fitness'].mean()
mean
```

```python
#piechart
fig=px.pie(data, values='Mental_fitness',names='Year')
fig.show()
```

```python
#creating new dataset by cpoying the older ones
dc=data.copy()
dc.head(5)
```

```python
dc.shape
```

```python
#non numeric lables to numeric lables
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in dc.columns:
  if dc[i].dtype == 'object':
     dc[i]=l.fit_transform(dc[i])
```

```python

#splitting data for training and test
from sklearn.model_selection import train_test_split
x=dc.drop('Mental_fitness',axis=1)
y=dc['Mental_fitness']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.20,random_state=2)

```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
#applying algorithms
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
lr=linear_model.LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_train_predict=lr.predict(x_train)
```

```python
# average of square of difference between observed and predicted values using root mean square erro
mse= mean_squared_error(y_train, y_train_predict)
```

```python
# Average difference between observed and predicted values using root mean square error
rmse=np.sqrt(mean_squared_error(y_train,y_train_predict))
```

```python
#coefficient of Determination describes how good a line is fit
r2=r2_score(y_train,y_train_predict)
```

```python
print("All the obtained data ")
print('MSE {} '.format(mse))
print('RMSE {} '.format(rmse))
print('R2 {} '.format(r2))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
y_train_predict=rf.predict(x_train)
mse=mean_squared_error(y_train,y_train_predict)
rmse=np.sqrt(mean_squared_error(y_train,y_train_predict))
r2=r2_score(y_train,y_train_predict)
```

```python

print("All the obtained data ")
print('MSE {} '.format(mse))
print('RMSE {} '.format(rmse))
print('R2 {} '.format(r2))
```

```python
#test dataset evalution for linear regression
y_test_predict=lr.predict(x_test)
mse= mean_squared_error(y_test, y_test_predict)
rmse=np.sqrt(mean_squared_error(y_test,y_test_predict))
r2=r2_score(y_test,y_test_predict)

print("All the obtained data ")
print('MSE {} '.format(mse))
print('RMSE {} '.format(rmse))
print('R2 {} '.format(r2))
```

```python
#test dataset evaluation for random forest regression
y_test_predict=rf.predict(x_test)
mse= mean_squared_error(y_test, y_test_predict)
rmse=np.sqrt(mean_squared_error(y_test,y_test_predict))
r2=r2_score(y_test,y_test_predict)

print("All the obtained data ")
print('MSE {} '.format(mse))
print('RMSE {} '.format(rmse))
print('R2 {} '.format(r2))
```
